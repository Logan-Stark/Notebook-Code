```
# Reading in data in Pandas

df = pd.read_csv('Video_Games_Sales_as_at_22_Dec_2016.csv')
df.shape
```

```
# From a brief look at our data we can see we have NaN values
# our year of release also seems to be a float, we will fix this
# our data is also a mix of sales and ratings 
df.head()
```

## Cleaning dataset and creating subsets

```
# Making new data set for games and sales
df_gs = df.drop('Critic_Score', axis =1) 
df_gs = df_gs.drop('Critic_Count', axis =1) 
df_gs = df_gs.drop('User_Score', axis =1) 
df_gs = df_gs.drop('User_Count', axis =1) 
df_gs = df_gs.drop('Rating', axis =1) 
df_gs = df_gs.drop('Developer', axis =1) 
```

```
# Dropping Null values
# casting year to int
# we are dropping null values here because most old games did not have
# Critic scores, they do however have sales numbers.

df_gs = df_gs.dropna( how='any')
df_gs['Year_of_Release'] = df_gs['Year_of_Release'].astype(int)
```

```
#checking changes
df_gs
```

```
# Dropping Null values
# casting year to int

df_rating = df
df_rating = df_rating.dropna(how= 'any')
df_rating['Year_of_Release'] = df_rating['Year_of_Release'].astype(int)
df_rating
```

## Feature and subset Engineering

```
# Adding the Column Rank to the front of our dataframe
# Our data frame is already in descending order so a simple column like this will allow us to call 
# games by there rank in cells

df_gs['Rank'] =  np.arange(1,16417)
```

```
# setting column as index to move it to far left

df_gs = df_gs.set_index('Rank')
```

```
# resetting our index to make rank our first column

df_gs = df_gs.reset_index('Rank')
```

```
# Creating a column for standardized regional sales

df_gs['NA_Sales_Adj'] = df_gs['NA_Sales']/579
```

```
# Creating a column for standardized regional sales

df_gs['JP_Sales_Adj'] = df_gs['JP_Sales']/126.5
```

```
# Creating a column for standardized regional sales

df_gs['EU_Sales_Adj'] = df_gs['EU_Sales']/741.4
```

```
# creating a subset for the top 100 best selling games for easier graphing 
# this is mostly because if we were to graph the entire dataset it would become very large
# a Large dataset is a hassle to get into our blog later

top100 = df_gs[:100]
```

## Standardized (North America, Europe, and Japan sales of the top 100 Video games)
- This grap took me the longest to make. it was my first plotly graph so it took me some time to figure out the basics. The graph show the relationship betwen the Regional sales and the top 100 selling video games using standardized data.

```
# figure
fig = go.Figure()

# Variables to be used later
Rank = top1000['Rank']
Name = top1000['Name']
NA_Sales = top1000['NA_Sales_Adj']
EU_Sales = top1000['EU_Sales_Adj']
JP_Sales = top1000['JP_Sales_Adj']
OT_Sales = top100['Other_Sales']

# Data for our figure (traces represent Region sales)
data = (fig.add_trace(go.Scatter(x=Rank, y=NA_Sales,
                    mode='lines+markers',
                    name='North America',
                    text= Name,
                    marker= dict(color = 'Dark cyan',size=8))),
fig.add_trace(go.Scatter(x=Rank, y=EU_Sales,
                    mode='lines+markers',
                    name='Europe', 
                    text= Name,
                    marker= dict(color = 'orange',size=8))),
fig.add_trace(go.Scatter(x=Rank, y=JP_Sales,
                    mode='lines+markers',
                    name='Japan',
                    text= Name,
                    marker= dict(color = 'red',size=8))))

# Figure layout (Title, Labels, etc)
fig.update_layout(dict(title = 'North America, Europe, and Japan Sales per person of Top 100 Video Games',
              xaxis= dict(title= 'Rank',ticklen= 5,zeroline= False,zerolinewidth=1,gridcolor="white"),
              yaxis= dict(title= 'Sales(Per Person)',ticklen= 5,zeroline= False,zerolinewidth=1,gridcolor="white",),
              paper_bgcolor='rgb(243, 243, 243)',
              plot_bgcolor='rgb(243, 243, 243)' ))

# Hover text grouping
fig.update_layout(hovermode="x")

# Figure Annotations
fig.update_layout(annotations=[dict(
            x=-.075,
            y=-0.27,
            showarrow=False,
            text="Stark Data",
            xref="paper",
            yref="paper"),
        dict(
            x=-0.075,
            y=-0.15,
            showarrow=False,
            text="______________________________________________________________________________________________________________________________________________________________________________________________",
            xref="paper",
            yref="paper"
        ),
        dict(
            x=1.15,
            y=-0.27,
            showarrow=False,
            text="SOURCE:VGCHARTS.COM",
            xref="paper",
            yref="paper"
        )],
    autosize=True,
    margin=dict(
        b=100),
    xaxis=dict(
        autorange=False,
        range=[0,100],
        type="linear"
    ),
    yaxis=dict(
        autorange=False,
        range=[-.01,.1],
        type="linear"
    ),
    height=550,
    width=1137
)
# save graph for exporting
pio.write_html(fig, file='Std-top-100-sales.html', auto_open=True)

fig.show()
```

## Genre by the total number of regional sales
- This graph shows the relationship betwen number of regional sales and genre. This was made using our unstandardized data and was grouped by region.


```
# Variables
Rank = top1000['Rank']
Name = top1000['Name']
NA_Sales = top1000['NA_Sales']
EU_Sales = top1000['EU_Sales']
JP_Sales = top1000['JP_Sales']
OT_Sales = top1000['Other_Sales']

# Bar plots
fig = go.Figure(data=[

go.Bar(name='North America', x= top1000['Genre'], y=NA_Sales, hovertext=Name, marker=dict(color = 'dark cyan')),
go.Bar(name='Europe',x= top1000['Genre'], y=EU_Sales, hovertext=Name, marker=dict(color = 'orange')),
go.Bar(name='Japan',x= top1000['Genre'], y=JP_Sales, hovertext=Name, marker=dict(color = 'red'))
])

fig.update_layout(barmode='group')

# Adjusting margins and Graph size
fig.update_layout(margin=dict(b=100),
                  yaxis=dict(autorange=False,range=[-1,450],type="linear"),
                  height=450,
                  width=1100)

# Adding Text and Title
fig.update_layout(title = 'Trends in Genre by Region Sales')
fig.update_layout(annotations=[dict(
            x=-.075,
            y=-0.4,
            showarrow=False,
            text="Stark Data",
            xref="paper",
            yref="paper"),
        dict(
            x=-0.075,
            y=-0.22,
            showarrow=False,
            text="______________________________________________________________________________________________________________________________________________________________________________________________",
            xref="paper",
            yref="paper"),
        dict(
            x=1.17,
            y=-0.4,
            showarrow=False,
            text="SOURCE:VGCHARTS.COM",
            xref="paper",
            yref="paper"
        )], xaxis=dict(title= 'Genre',zeroline= False,zerolinewidth=0,gridcolor="white"),
            yaxis= dict(title= 'Sales(In Millions)',ticklen= 5,zeroline= False,zerolinewidth=1,gridcolor="white",))

# save graph for exporting
pio.write_html(fig, file='Regions-by-Genre.html', auto_open=True)

fig.show()

```
